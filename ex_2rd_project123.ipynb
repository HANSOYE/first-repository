# 1. 프로젝트 load_digits : 손글씨를 분류해 보기
## (1) 필요한 모듈 import하기

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

## (2) 데이터 준비

```python
digits = load_digits()

print(dir(digits))
```

```python
digits.keys()
```

```python
digits_data = digits.data

print(digits_data.shape)
```

```python
digits_data[0]
```

```python
#손글씨 데이터의 타겟 정보

digits_label = digits.target
print(digits_label.shape)
digits_label
```

```python
digits.target_names
```

```python
#데이터 Describe 해 보기
print(digits.DESCR)
```

## (3) 데이터 이해하기

```python
#feature에 대한 설명
digits.feature_names
```

```python
import pandas as pd

print(pd.__version__)
```

```python
#손글씨 데이터셋을 pandas가 제공하는 datapframe의 자료형으로 변환하기
#Feature Data 지정하기
digits_df = pd.DataFrame(data = digits_data, columns = digits.feature_names)
digits_df
```

```python
#Label Data 지정하기
digits_df["label"] = digits.target
digits_df
```

```python
#Target Names 출력해 보기
digits.target_names
```

```python
import numpy as np

digits_df.describe()
```

## (4) train, test 데이터 분리

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(digits_data, # 데이터 셋인 x
                                                    digits_label, # 정답인 y
                                                    test_size=0.2,  #전체의 20%를 테스트 데이터로 쓰겠다.
                                                    random_state=7) #train 데이터와 test데이터를 분리하는데 적용되는 랜덤성을  결정

print('X_train 개수: ', len(X_train), ', X_test 개수: ', len(X_test))
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
#y 확인
y_train, y_test
```

## (5) 다양한 모델로 학습시켜보기
# 

* Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

decision_tree = DecisionTreeClassifier(random_state = 32)
print(decision_tree._estimator_type)
```

```python
#학습시키기
decision_tree.fit(X_train, y_train)
```

```python
#학습 완료 후 X_test 데이터로 예측하여 y값 예측하기
y_pred = decision_tree.predict(X_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
accuracy
```

* Random Forest 사용하기

```python
from sklearn.ensemble import RandomForestClassifier

X_train, X_test, y_train, y_test = train_test_split(digits_data,
                                                    digits_label,
                                                    test_size=0.2,
                                                    random_state=21)

random_forest= RandomForestClassifier(random_state = 32)
random_forest.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)

print(classification_report(y_test, y_pred))
```

* SVM 모델 사용

```python
from sklearn import svm
svm_model =svm.SVC()

svm_model.fit(X_train, y_train)
y_pred = decision_tree.predict(X_test)

print(classification_report(y_test, y_pred))
```

* SGD Classifier 모델 사용하기

```python
from sklearn.linear_model import SGDClassifier
sgd_model = SGDClassifier()

print(sgd_model._estimator_type)

sgd_model.fit(X_train, y_train)
y_pred = sgd_model.predict(X_test)

print(classification_report(y_test, y_pred))
```

* Logistic Regression 모델 사용하기

```python
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression(max_iter=3000)

logistic_model.fit(X_train, y_train)
y_pred = logistic_model.predict(X_test)

print(classification_report(y_test, y_pred))
```

## (6) 모델을 평가해 보기

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred)
```

모든 데이터의 분석의 결과로 하나의 결론을 내는 것이 아닌, 각자의 숫자의 정확성을 판별하기만 하면 되는 종류의 데이터이다. 즉,  각 데이터 사이의 상호작용이 중요하지 않은 데이터이기 때문에 평가지표로 Accuracy를 선택했다.
Accuracy는 전체 대비 정확하게 예측한 개수의 비율로, 손글씨 분류하기에서는 테스트 이미지 중 올바르게 분류한 비율을 의미한다.

Accuracy를 살펴보면, 그 값들이 다 비슷하기 때문에 Random Forest, SVM, SGD Classifier, Logistic Regression 중, 그 어느 것을 사용해도 상관 없다.

Recall 을 평가지표로 사용하는 것이 타당하다.
Recall은 (TP)/(TP+FN)이다.
FN, 즉 양성인데 음성으로 판단 된 결과가 적을 수록 Recall 값이 크게 나온다.
유방암의 경우 양성인데 양성 판정, 음성인데 음성 판정의 경우에는 문제가 없다.
또한, 음성인데 양성으로 판단되어도 실제로 유방암에 걸린 것은 아니기에 크게 문제될 것은 없으나, 양성인데 음성 판정이 되어 환자가 제때 치료를 받지 못한다거나 하는 사고는 일어나지 않아야 한다.
그렇기 때문에 recall을 평가지표로 사용하는 것이 옳다.

Recall 값을 보면, Random Forest와 SVM의 Recall값이 높기 때문에, 유방암 판별에서는 이 두 모델을 사용하는 것이 옳다.
